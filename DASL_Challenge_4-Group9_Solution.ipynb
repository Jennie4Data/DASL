```python
#Rename specific columns
movies.rename(columns={'MOVIES': 'Movies', 'YEAR': 'Year', 'GENRE': 'Genre', 'RATING': 'Rating', 'ONE-LINE': 'One-Line', 'STARS': 'Stars',  'VOTES': 'Votes', 'Gross': 'Gross($)'}, inplace=True)

# Print the updated DataFrame
print(movies)
print(movies.columns)
```

#### Remove duplicate values from the movies column

```python
#Remove duplicate values from the 'movies' data instead, there are information in other column that are important, if analysis comes in later on
movies = movies.drop_duplicates(subset=['Movies'])

#Reset the index
movies = movies.reset_index(drop=True)

#Print the updated DataFrame
print(movies)
```

```python
#See the first few rows of the DataFrame
print(movies.head(10))
```

#### Clean the year column to get the appropriate year

```python
movies['Year'] = movies['Year'].astype(str)

# Clean the "Year" column to extract the appropriate year or year range
movies['Year'] = movies['Year'].str.replace('[^0-9-]', '', regex=True)  # Remove non-digit and non-dash characters
movies['Year'] = movies['Year'].apply(lambda x: re.sub(r'–', '-', x))  # Replace dash-like character with a hyphen

# Print the updated DataFrame

# Convert the 'year' column to the integer data type
movies['Year'] = pd.to_numeric(movies['Year'], errors='coerce').astype('Int64')

# Print the updated DataFrame
print(movies)
```

```python
# Convert the "Year" column to string
movies['Year'] = movies['Year'].astype(str)
# Clean the "Year" column to get the appropriate year or year range
movies['Year'] = movies['Year'].str.replace(r'(\d{4})(\d{4})', r'\1-\2', regex=True)

# Print the updated DataFrame
print(movies)
```

```python
# Display the first few rows of the DataFrame
print(movies.head(100))
```

#### Get the first genre from the genre column


```python
# Extract the first genre from the 'genre' column
movies['First Genre'] = movies['Genre'].str.split(',').str[0].str.strip()

# Remove any leading/trailing whitespace and convert to lowercase
movies['First Genre'] = movies['First Genre'].str.lower()

#Remove white space in genre cells
movies['Genre'] = movies['Genre'].str.strip()

# Print the updated DataFrame
print(movies)
```

#### Clean the Rating column


```python
#Replace null values in the 'Rating' column with a default value of 'Not Rated' assuming that those nulls cells means that they were not rated in the first place.
movies['Rating'] = movies['Rating'].fillna('Not Rated')

# Print the updated DataFrame
print(movies)
```

#### Get the director name and put it in a different column and also put the stars names in a separate column


```python
# Convert non-string values in the 'Stars' column to strings
movies['Stars'] = movies['Stars'].astype(str)

# Extract the director name and place it in a separate column
movies['Director'] = movies['Stars'].str.split('Director:').str[1].str.split('|').str[0].str.strip()

# Extract the stars' names and place them in a separate column
movies['Starring'] = movies['Stars'].str.split('Stars:').str[1].str.split(',').str.join(',').str.strip()

# Remove newline characters from the stars' names
movies['Starring'] = movies['Starring'].str.replace('\n', '')

# Display the modified DataFrame
print(movies[['Director', 'Starring']])
print(movies)
```

```python
# Replace null values in the "Director" and "Starring" column with a default value of 'Unknown' assuming, they are unknown.
movies['Director'] = movies['Director'].fillna('Unknown')
movies['Starring'] = movies['Starring'].fillna('Unknown')

# Print the updated DataFrame
print(movies[['Director', 'Starring']])
```

```python
#Apply the unidecode function to the  Director and Starring column to replace irregular lettering
movies['Director'] = movies['Director'].apply(lambda x: unidecode(x))
movies['Starring'] = movies['Starring'].apply(lambda x: unidecode(x))

#Dropping the Stars since we have a sepearte column for director- 'Director' and stars- 'Starring'
movies = movies.drop('Stars', axis=1)
```

#### Clean the votes column


```python
#Convert the 'votes' column to numeric data type
movies['Votes'] = pd.to_numeric(movies['Votes'], errors='coerce')

#Assuming votes gotten for a particular movie is based on the genre of that movie and also that people tend to watch a particular genre more than the other.
#We fill the null cells with the mean of others movies in it's genre.
#The votes a movie receives can be influenced by various factors, and the genre is one of them. 
#Disclaimer:it is important to note that while genre can influence the voting patterns, it is not the sole determining factor- due to lack of further information, that can not be determined.
movies['Votes'] = movies.groupby('First Genre')['Votes'].transform(lambda x: x.fillna(x.mean()))

#Print the updated DataFrame
print(movies)

```

#### Clean the runtime column


```python
# Remove non-numeric characters and convert the 'runtime' column to numeric data type
movies['RunTime'] = pd.to_numeric(movies['RunTime'], errors='coerce')

#Runtime is basically,In the context of movies, refers to the duration or length of a film from its beginning to its end. 
#So, we assume that it is supposed to be a time duration- in minutes
# Extract the hours and minutes from the runtime column
movies['hours'] = movies['RunTime'] // 60
movies['minutes'] = movies['RunTime'] % 60

# Convert the runtime column to minutes
movies['RunTime'] = movies['hours'] * 60 + movies['minutes']

# Remove the intermediate 'hours' and 'minutes' columns if needed
movies = movies.drop(['hours', 'minutes'], axis=1)

#Print the updated DataFrame
print(movies)
```

```python
#Upon filling null cells, we assume the same thing as voting, where we used the mean of the runtime of similar genre.
# Calculate the mean runtime for each genre
genre_mean = movies.groupby('First Genre')['RunTime'].mean()

# Fill null values in the runtime column with the mean of the corresponding genre
movies['RunTime'].fillna(movies['First Genre'].map(genre_mean), inplace=True)

#Print the updated DataFrame
print(movies)
```

#### Clean the gross column

```python
# Remove non-numeric characters and convert the 'runtime' column to numeric data type
movies['Gross($)'] = movies['Gross($)'].astype(str).str.replace('[^0-9]', '', regex=True)
movies['Gross($)'] = pd.to_numeric(movies['Gross($)'], errors='coerce')

#Following the same assumption that null cells will be filled by the mean of the gross based on the genre
#Becausethe genre of a movie can be a factor that influences its gross revenue.
#Different genres tend to attract different audiences and have varying levels of commercial appeal.
genre_mean = movies.groupby('First Genre')['Gross($)'].transform('mean')

# Fill null values in the 'gross' column with the mean of the corresponding genre
movies['Gross($)'].fillna(genre_mean, inplace=True)

print(movies)
```

```python
# Round the values in 'Votes' to the nearest integer
movies['Votes'] = movies['Votes'].round()

# Round the values in 'RunTime' to 2 decimal places
movies['RunTime'] = movies['RunTime'].round(2)

# Round the values in 'Gross($)' to 2 decimal places
movies['Gross($)'] = movies['Gross($)'].round(2)
```

```python
#Remove White Space in the One-line Column
movies['One-Line'] = movies['One-Line'].str.strip()
```
